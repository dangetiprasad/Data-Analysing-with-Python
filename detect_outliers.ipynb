# User defined function to detect and remove outliers in any dataset


```
def detect_outlier(data):
    b=data.columns
    for i in b:
        if data[i].dtypes=='float':
            mean=np.mean(data[i])
            std=np.std(data[i])
            threshold=3
            lkj=[]
            for j in data[i].values:
                z_score=(j-mean)/std
                if np.abs(z_score)>threshold:
                    lkj.append(j)
                
            data[i]=data[i].replace(lkj,mean)     
   
    return data
#taken datase2 as input dataset    
detect_outlier(a3)
```
